### Ejercicio 6
##### Apartado 1

```python


def elegirParticula(n):
    return randint(0, n-1)

def simular(n, N):
    L = [5 for _ in xrange(n)]
    for t in xrange(N):
        j = elegirParticula(n)
        i = j
        while j == i:
            i = elegirParticula(n)
            
        if L[i] > 0:
            L[i] -= 1
            L[j] += 1
    return L

simular(10, 10^5)
```

### Ejercicio 9
##### Apartado 1

```python
import random as r

def paso():
    L = []
    n = randint(0, 3)
    if n == 0:
        L = [1, 0]
    elif n == 1:
        L = [0, 1]
    elif n == 2:
        L = [0, -1]
    elif n == 3:
        L = [-1, 0]
    return L
    
    
def distancia(pasos, N):
    L = [0, 0]
    for i in xrange(0, N):
        for k in range(pasos):
            p = paso()
            L[0] += p[0]
            L[1] += p[1]
            
    L[0] = (L[0]/N).n()
    L[1] = (L[1]/N).n()
    return L

distancia(1000, 1000)
```

##### Apartado 2

```python
def vuelveCentro(pasos, N):
    n = 0
    for i in xrange(0, N):
        L = [0, 0]
        for k in range(pasos):
            p = paso()
            L[0] += p[0]
            L[1] += p[1]
            if L.count(0) == 2:
                n += 1
                break
    return (n/N).n()

%time vuelveCentro(1000, 10000)
```

##### Apartado 3

```python
def coincidenPaseantes(pasos, N):
    n = 0
    for i in xrange(0, N):
        histL1 = []
        L1 = [0, 0]
        L2 = [0, 0]
        for k in range(pasos):
            p = paso()
            L1[0] += p[0]
            L1[1] += p[1]
            histL1.append(L1)
            p = paso()
            L2[0] += p[0]
            L2[1] += p[1]
            if L2 in histL1:
                n += 1
                break
                
    return (n/N).n()

%time coincidenPaseantes(1000, 1000)
```

##### Apartado 4

```python
def coincidenPaseantes2(pasos, N):
    n = 0
    for i in xrange(0, N):
        L1 = [0, 0]
        L2 = [0, 0]
        for k in range(pasos):
            p = paso()
            L1[0] += p[0]
            L1[1] += p[1]
            p = paso()
            L2[0] += p[0]
            L2[1] += p[1]
            if L1[0] == L2[0] and L1[1] == L2[1]:
                n += 1
                break
                
    return (n/N).n()

coincidenPaseantes2(10000, 1000)
```

### Ejercicio 10
##### Apartado 1

```python
def cambiarUrna(Ut):
    color = randint(0, sum(Ut))
    if(color <= Ut[0]):
        Ut[0] += 1
    else:
        Ut[1] += 1
    #Devolvemos una copia de la lista
    return list(Ut)
```

##### Apartado 2

```python
def simularUrna(N):
    L = []
    U = [0, 0]
    for _ in xrange(N):
        U = cambiarUrna(U)
        L.append(U)
    return L

L = simularUrna(10000)
Puntos = []
for i in range(len(L)):
    pBlancas = (L[i][0]/sum(L[i])).n()
    c = [i, pBlancas]
    Puntos.append(c)
    
line2d(Puntos)
```
