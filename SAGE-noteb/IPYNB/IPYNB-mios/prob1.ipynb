```python
def moneda_trucada(k):
    x = random()
    if x<= k:
        return 0
    return 1

def caras_100(k):
    cont = 0
    for muda in xsrange(1000):
        x  = moneda_trucada(k/10)
        if x == 0:
            cont = cont + 1
        if cont > 100:
            return 0
    return 1
```

```python
def ejercicio_1(k):
    cont = 0
    for i in xsrange(10000):
        x = caras_100(k)
        if x == 1:
            cont = cont + 1
    return (cont/10000).n()
```

```python
%time ejercicio_1(1)
```

```python
#Ejercicio
def particulos(n,t):
    L = [5 for dani in xsrange(n)]
    for i in xsrange(t):
        a = randint(0,n-1)
        b = randint(0,n-1)
        while a == b:
            b = randint(0,n-1)
        if L[a] != 0 :
            L[a] = L[a] - 1
            L[b] = L[b] + 1
    return L
```

```python
L = particulos (5,10)
```

```python
[finance.TimeSeries(particulos(1000,10^n)).plot_histogram(bins = 100,normalize = False) for n in xsrange(1,7)]
```

```python
#Ej Paseos
def paso(L):
    p = randint(0,4)
    if p == 0:
        L[0] = L[0] + 1
    if p== 1:
        L[0] = L[0] - 1
    if p == 2:
        L[1] = L[1] + 1
    if p == 3:
        L[1] = L[1] - 1
    return L

def dar_paseo(N):
    L = [0,0]
    for i in xsrange(N):
        paso(L)
    modulo = sqrt(L[0]^2 + L[1]^2).n()
    return modulo,L
```

```python
dar_paseo(1000)
```

```python
def pasa_centro(N,rep):
    L = [0,0]
    res = 0
    for i in xsrange(rep):
        for j in xsrange(N):
            paso(L)
            if L[0] == 0 and L[1] == 0:
                res = res + 1
                break
    return res
```

```python
pasa_centro(1000,10000)
```

```python
#Polya
#U = [numblancas,numnegras]
def U(L):
    nummax = L[0] + L[1]
    bola = randint(1,nummax)
    if bola <= L[0]:
        L[0] += 1
        return L
    L[1] += 1
    return L
```

```python
def urnas_polya(N):
    L = [1,1]
    res = list()
    res.append((L[0],L[1]))
    for int in xsrange(N):
        L = U(L)
        res.append((L[0],L[1]))
    return res
```

```python
urnas_polya(1000)
```
