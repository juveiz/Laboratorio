Prueba de algoritmo recursivo con grafo que dibuje los nodos.

```python
def potencia(n, k, dicc):
    if k == 0:
        dicc[k]=[0]
        return 1
    if k == 1:
        dicc[k]=[n]
        return n
    if k%2 == 0:
        pot = potencia(n, k//2, dicc)
        dicc[k]=[pot, pot]
        return pot*pot
    if k%2 == 1:
        pot = potencia(n, (k-1)//2, dicc)
        dicc[k]=[pot, pot]
        return pot*pot*n
    
dicc = {}
potencia(2, 5, dicc)

Graph({0:[1,2,3], 2:[4]})
```

## Listas

```python
L=[1,2,3,4]

#Añade la lista entera a L
L.extend([8,9,10])

print(L)
#Cuenta el número de apariciones
L.count(1)
#Indice de la primera vez que aparece
L.index(1)

#Elimina la primera aparicion del valor
L.remove(1)
print(L)
#Elimina el elemento de la posicion 1
del L[1]
print(L)
#Elimina el elemento de la posicion y lo devuelve
L.pop(1)
print(L)

#Inserta en la posicion 2 el 18
L.insert(2, 18)
print(L)

#Vaciar una lista
del L[:]
print(L)
```

```python
N=12758160859749830187568

#Número de cifras decimales
print floor(log(N, base=10))+1

# P = 28433*2^(7830457)+1
#Últimas 10 cifras por la derecha. Solo nos importa el resultado en módulo 10^10
print (28433*power_mod(2,7830457,10^10)+1)%10^10

#Cifras dominantes de un número k^n
def cifra_dom(k,n):
    '''Cifra dominante de k^n. Ver 7.5 de las notas del curso'''
    return floor(10^((n*log(k,base=10).n())-((n*log(k,base=10).n()).floor())))

def m_cifras_dom(m,k,n):
    '''m cifras mas dominantes de k^n. Ver 7.5 de las notas del curso'''
    return floor(10**m*10^((n*log(k,base=10).n())-((n*log(k,base=10).n()).floor())))

cifra_dom(4,2)
```
