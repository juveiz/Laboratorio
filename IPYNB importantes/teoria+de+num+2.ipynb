```python
#Ejercicio 9
def wilson(n):
    aux = [True]*int(n)
    aux[0] = False
    aux[1] = False
    for i in xrange(2,floor(sqrt(n))+1):
        if aux[i]:
            for j in xrange(i*i,n,i):
                aux[j] = False
    return [k for k in xrange(n) if aux[k]==True]
```

```python
def primos(n):
    L=list()
    for i in xsrange(2,n+1):
        flag=0
        for j in xsrange (2,floor(sqrt(i))+1):
            if i%j == 0:
                flag=1
        if flag== 0:
            L.append(i)
    return L
```
