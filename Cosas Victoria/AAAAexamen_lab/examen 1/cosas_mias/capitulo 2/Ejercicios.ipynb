## Ejercicio 1. Demuestra por inducción sobre n ∈ N 

## a

```python
#sum(expresion, variable, limite inferior, limite superior)
def suma_cuadrados(n):
    var('res', 'i')
    res = sum(i^2, i, 1, n)
    return res
```

```python
suma_cuadrados(3)
```

```python
def hipotesis(i):
    return (i * (i + 1)*(2*i + 1)) / 6
def induccion(k):
    L = []
    for x in srange(1, k):
        if hipotesis(x) == suma_cuadrados(x):
            L.append(1)
        else:
            L.append(0)
    return L
```

```python
all(induccion(10))
```

## b

```python
def suma_fracciones(n):
    suma=0
    for i in srange(1,n+1):
        suma+=1/(i*(i+1))
    return suma
```

```python
suma_fracciones(2)
```

```python
def hipotesis_1(i):
    return (i/(i+1))
def induccion_1(k):
    L = []
    for x in srange(1, k):
        if hipotesis_1(x) == suma_fracciones(x):
            L.append(1)
        else:
            L.append(0)
    return L
```

```python
#probamos para 1
all(induccion_1(1)), all(induccion_1(1000)) #probamos para un n mayor
```

```python
def fibonachi(num):
    a,b=0,1
    for i in srange(1,num+1):
        a,b=b,a+b
        #print a
    return a
```

```python
fibonachi(8)
```

```python
def sum_fibonachi(n):
    sum=0
    for i in srange(1,n+1):
        f=fibonachi(i)
        sum+=f
        #print sum
    return sum
```

```python
sum_fibonachi(3)
```

```python
def hipotesis_f(n):
    return (((((1+sqrt(5))/2)^n) - ((1-sqrt(5))/2)^n)/sqrt(5))
def induccion_f(k):
    L = []
    for x in srange(1, k):
        if hipotesis_f(x) == sum_fibonachi(x):
            L.append(1)
        else:
            L.append(0)
    return L
```
