```python
m=5
[derivative(x^m,x,j) for j in range(m+2)]
```

```python
F(x,y)=x^2*sin(y)+y^2*cos(x)
#Derivamos F en duncion de x
show([derivative(F,x,1)]);
#Derivamos F en duncion de y
show([derivative(F,y,1)]);
#derivamos F en funcio de y y luego de y
show([F.derivative(y,2)])
#Derivamos F en funcio de x y luego de y
show([derivative(F,x,1).derivative(y,1)])
```
