plot muestra la gr√°fica de p evaluado, entre los valores definidos en el segundo argumento.

```python
p(x,a,b) = x^4 - 6 * x^2 + a*x + b

```

```python
show(p)
```

```python
plot(p(x,0,0), (x, -3,3))
```

```python
plot(p(x,0,2), (x, -3,3))
```

```python
plot(p(x,0,-9), (x, -3,3))
```

```python
plot(p(x,0,99), (x, -3,3))
```

```python
plot3d(p(x,0,b), (x,-3,3), (b,-100,100))
```

```python
plot(p(x,8,0), (x, -3,3))
```

```python
show(p(x,0,b).factor())
```

```python
solve(p(x,0,b),x)
```
