$\dfrac{1}{1\cdot 2}+\dfrac{1}{2\cdot 3}+\cdots+\dfrac{1}{n\cdot (n+1)} =
\dfrac{n}{n+1}\,,\text{ si }n\ge1.$

```python
def resultado(n):
    return n/(n+1)
```

```python
def sumadefracciones(n):
    suma = 0
    i = 1
    while i <= n:
        suma += 1/(i*(i+1))
        i += 1
    return suma
```

```python
L=[]
i=1
while i < 1000:
    L.append(sumadefracciones(i) == resultado(i))
    i+=1
```

```python
all(L)
```
