# Sucesi√≥n de Fibonacci

```python
var('epsilon')
epsilon = 0.001
```

```python
def fibonacci(n):
    aux1=0
    aux2=1
    aux3=0
    i = 2
    while i < n:
        aux3 = (aux1 + aux2)
        aux1 = aux2
        aux2 = aux3
        i += 1
    return aux1 + aux2
```

```python
def resultado(n):
    return(((((1+sqrt(5))/2)^n) - ((1-sqrt(5))/2)^n)/sqrt(5))
```

```python
fibonacci(100)
```

```python
resultado(100)
```

```python
L = []
for i in srange(1, 1000):
    L.append(abs(fibonacci(i) - resultado(i)).n() <= epsilon )
```
