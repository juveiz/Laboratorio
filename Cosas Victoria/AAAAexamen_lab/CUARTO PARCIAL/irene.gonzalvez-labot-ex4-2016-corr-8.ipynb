<p>la funcion elemento devuelve un elemento aleatorio de una lista</p>

```python
def elemento(L):
    n=len(L) - 1
    i=randint(0,n)
    return L[i]
```

<p>la funcion mancha devuelve la porcion de tela (numerada) por la gota de tinte</p>

```python
def mancha(t,k):
    N=(2*k+1)*t
    L=srange(k,N-k +1)
    x=elemento(L)
    LL=srange(x-k, x+k+1)
    return LL
```

```python
mancha(5,2)
```

<p>esta funcion analiza si dos manchas se solapan</p>

```python
def solapar(t,k):
    # L1 y L2 son de la misma longuitud
    L1=mancha(t,k)
    L2=mancha(t,k)
    for j in srange(len(L1)):
        for m in srange(len(L1)):
            if L1[j]==L2[m]:
                return 1

    return 0
```

```python
def pr_solapen(t,k,M):
    L=[solapar(t,k) for muda in srange(M)]
    #M es el numero de veces que realizamos el experimento de que dos gotas se solapen
    p=(sum(L)/M).n()
    return p
```

<h4>Apartado 2</h4>

<p>voy a crear una funcion que en una unica lista esten el numero de los trozos de tejido te&ntilde;idos y me devuelva si estan el tejido cubierto por el tinte</p>

```python
def tintes(t,k,s):
    L=[]
    for j in srange(s):
        LL=mancha(t,k)
        for i in srange(2*k+1):
            L.append(LL[i])
    cubierto =list(set(L))
    n=len(cubierto)
    
    if n==(2*k + 1)*t + 1:
        return 1
    else:
        return 0
```

```python
tintes(5,2,10)
```

```python
def pr_cubrir(t,k,s,M):
    L=[tintes(t,k,s) for muda in srange(M)]
    p=(sum(L)/M).n()
    return p
```

```python
pr_cubrir(5,2,9,10^3)
```

```python
def f(x,a):
    y=x^a
    return y
```

<p>calculo de la integral mediante el area</p>

```python
def cal_area(a,N):
    # N es el numero de 'dardos' lanzados
    contador=0
    for muda in srange(N):
        x=random()
        y=random()
        if f(x,a)>=y:
            contador = contador + 1
    return (contador/N).n()
```

<p>mediante sumas de Riemann</p>

<p>azar elige un numero de forma aleatoria en el intervalo [a,b]</p>

```python
def azar(a,b):
    x=random()
    return a + x*(b-a)
```

```python
def cal_R(a,N):
    # el intervalo [1,0] dividido en N subintervalos de la misma longuitud
    L=[]
    h=1/N
    for j in srange(N+1):
        a=h*j
        L.append(a)
    LL=[]
    for i in srange(N):
        x=azar(L[i],L[i+1])
        y=f(x,a)
        LL.append(y)
    integral=sum(LL)
    return integral
```
