<h3>Ruina del jugador</h3>

Si sale cara le paga un euro el casino, y si no le paga él un euro al casino



```python
def moneda():
    x = random()
    if x <= 0.5:
        return -1
    else:
        return 1
```

```python
def ruina(N): ## N es la riqueza en el instante 0 del jugador
    L = [N]
    while N > 0: ## No es un bucle infinito dado que el jugador cuando se quede sin dinero, saldrá
        N += moneda()
        L.append(N) ## Guarda la riqueza del jugador en el tiempo
    return L
```

```python
L=ruina(100); len(L) ## Termina cuando se arruina, pero lo que devuelve es el tiempo que ha podido estar jugando el jugador
```

```python
LL=zip(srange(len(L)),L) ##zip es una función que de dos lista hace una lista de pares
```

len(L)) -> tiempo que ha estado jugando ///

```python
line(LL).show() ##unimos los puntos para ver la oscilación de la riqueza del jugador a lo largo del tiempo
```

```python
def ruina_len(N):
    cont = 1
    while N > 0:
        N += moneda()
        cont += 1
    return cont
```

<h3>Jugador precavido</h3>
<p>AHORA fijamos un tope superior, es decir el jugador abandona cuando llega a un cierto valor de su 'riqueza':</p>

```python
def ruina_s(N,M): ##El limite es M, cuadno llegue a M se plantará
    #L = [N]
    while M > N > 0:
        N += moneda()
    #   L.append(N)
    return N,M
```

```python
ruina_s(100,1100)
```

```python
def prob_no_ruina(N,M,R): ##probabilidad de que cuando se retire no es por estar arruinado sino por haber llegado a su límite
    cont = 0
    for int in srange(R):## R número de veces que se repetirá el proceso, número arbitrario no demasiado grande
            ruina,exito=ruina_s(N,M)
            if ruina != 0:
                cont += 1
    return (cont/R).n()
```

```python
prob_no_ruina(100,110,10000)
```

```python
LL=[(100+k*10,prob_no_ruina(100,100+k*10,1000)) for k in srange(10)]
```

<h3>Casino limitado</h3>

Puede ser que el jugador rompa la banca

```python
def casino_limitado(RC,RJ): ##Riqueza casino, riqueza jugador
    while (RC>0 and RJ>0):
        x = moneda()
        if x == 1:
            RJ += 1
            RC -= 1
        else:
            RJ -= 1
            RC += 1
    return RC,RJ
```

```python
casino_limitado(200,100)
```

```python
def prob_ruina_jugador(RCI,RJI,R):
    cont = 0
    for muda in xrange(R):
        rcasino,rjugador = casino_limitado(RCI,RJI)
        if rjugador == 0:
            cont += 1
    return (cont/R).n()
```

```python
prob_ruina_jugador(100,100,1000) ## tienen el mismo dinero y por eso la probabilidad de que el jugador gane es de 1/2, cuanto más aumentemos el R más se aproximará
```

```python
prob_ruina_jugador(200,100,1000) ## La probabilidad de que el casino arruine el jugador aumenta
```
