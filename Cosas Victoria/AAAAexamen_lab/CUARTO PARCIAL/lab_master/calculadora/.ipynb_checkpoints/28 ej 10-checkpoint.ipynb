Comenzando con una base $B=\{v_1, v_2, v_3\}$

Donde $v_1, v_2 \in Ker(f)$

Para $v_3$ decido utilizar $(1, -1, 1)$, donde s√© que $f(v_3) = (-1, 1)$

```python
E = VectorSpace(QQ,3)
V3 = E([1,-1,1])
```

```python
Vn = E([1,1,-2])
```

Para hallar los vectores $v_1$ y $v_2$ obtengo tres puntos del plano, con los que formo dos vectores. Busco si esos vectores son perpendiculares al vector normal del plano y son linealmente independientes. En caso de que ademas sean linealmente independientes con $v_3$ tendremos una base de nuestro espacio

```python
P1 = E([0,0,0])
P2 = E([1,1,1])
P3 = E([2,0,1])
```

```python
V1 = P2 - P1
V2 = P3 - P1
```

```python
V1, V2
```

```python
A = matrix([V1, V2])
```

```python
A
```

```python
A.rank()
```

Como el rango es 2, son linealmente independientes

```python
V1.dot_product(Vn)
```

```python
V2.dot_product(Vn)
```

Viendo que son perpendiculares, veamos si son linealmente independientes todos los vectores de la base

```python
B = matrix([V1, V2, V3])
```

```python
B
```

```python
B.rank()
```

Como el rango es 3, son linealmente independientes, lo que provoca que tengamos una base completa, ahora escribamos la matriz respecto a esta base:

```python
C = column_matrix([[0,0], [0,0], [-1,1]])
```

```python
show(C)
```

Ahora cambiemos esa matriz a la base canonica

```python
MBC = column_matrix([[1,1,1], [2,0,1], [1,-1,1]])
```

```python
show(MBC)
```

```python
show(C*MBC)
```

Por lo que esa es nuestra matriz deseada
