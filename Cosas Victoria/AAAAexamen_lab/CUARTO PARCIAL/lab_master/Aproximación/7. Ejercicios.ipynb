# Ejercicio 5 Salamin y Brent

```python
def syb(n):
    a=1
    b=sqrt(2)/2
    s=1/2
    i=1
    while i<=n:
        ap=(a+b)/2
        bp=sqrt(a*b)
        a=ap
        b=bp
        c=a^2-b^2
        s=s-2^i*c
        p=(2*a^2)/s
        i=i+1
    return p.n()
```

```python
syb(10)
```

# Ejercicio 7 Rabinowitz y Wagon

```python
def pi_ryw(n):
    k=0
    for i in srange(n):
        a=((factorial(i))^2) * 2^(i+1)
        b=factorial(2*n+1)
        k+=a/b
    return k
```

```python
(pi_ryw(50).n(digits=10))
```

# Ejercicio 8 Log

```python
def log_apr(ndigits):
    k=1
    r=0
    while k <= ndigits:
        h=1/(k*2^k)
        r=r+h
        k+=1
    return r.n()
```

```python
log_apr(10)
```
