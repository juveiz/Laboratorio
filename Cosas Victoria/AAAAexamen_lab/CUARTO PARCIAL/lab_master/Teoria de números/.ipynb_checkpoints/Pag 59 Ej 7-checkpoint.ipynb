```python
def sumadecuadrados(N):
    L=[]
    flag = 0
    for i in prime_range(N+1):
        for x in srange(1, N):
            for y in srange(1, N):
                if(i == x**2 + y**2):
                    L.append(i)
                    flag = 1
                    break
            if(flag == 1):
                flag = 0
                break
    return L
```

Cualquier número que sea multiplicación de dos números que se puedan expresar como dos cuadrados será una suma de dos cuadrados, por lo que cualquier número cuyos factores se puedan expresar como suma de dos cuadrados será una suma de dos cuadrados

```python
def sumadecuadrados1(N):
    L=[]
    flag = 0
    for i in srange(N+1):
        for x in srange(-N, N):
            for y in srange(-N, N):
                if(i == x**2 + y**2):
                    L.append(i)
                    flag = 1
                    break
            if(flag == 1):
                flag = 0
                break
    return L
```

```python
sumadecuadrados1(80)
```
