```python
def N(x, y):
    return 10*x + y
```

```python
def F(x,y):
    return x - 2*y
```

```python
def multiplo_de_7(x):
    if(x%7 == 0):
        return True
    else:
        return False
```

```python
L = []
for x in srange(0, 100):
    for y in srange (0,10):
        L.append(multiplo_de_7(N(x,y)) and multiplo_de_7(F(x,y)))
L
```

```python
all(L[::7]) #comprobar que los múltiplos de 7 son True
```

```python
T = []
for x in srange(0, 100):
    for y in srange (0,10):
        T.append(multiplo_de_7(N(x,y)) and not multiplo_de_7(F(x,y)))
any(T)
```

```python
def fact(x):
    L = []
    L.append(x//10)
    L.append(x%10)
    return L
```

```python
fact(-14)
```

```python
def orbita(ini):
    P = []
    while not(ini in P):
        P.append(ini)
        ini = F(fact(ini)[0], fact(ini)[1])
    P.append(ini)
    return P
```

```python
orbita(2560)
```
